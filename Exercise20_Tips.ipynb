```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

% matplotlib inline
```

```python
df = pd.read_csv('https://github.com/guipsamora/pandas_exercises/raw/master/07_Visualization/Tips/tips.csv')
df.head()
```

```python
df.shape
```

```python
#Delete the Unnamed 0 column
df = df.iloc[:,1:]
df.head()
```

```python
df.shape
```

```python
#Plot the total_bill column histogram
plt.hist(df.total_bill)
```

```python
g = sns.distplot(df.total_bill);
g.set(xlabel = 'Value', ylabel = 'Frequency', title = "Total Bill")
```

```python
#Create a scatter plot presenting the relationship between total_bill and tip
plt.scatter(df.total_bill, df.tip)
```

```python
sns.jointplot(x ="total_bill", y ="tip", data = df)
```

```python
#Create one image with the relationship of total_bill, tip and size
sns.pairplot(df)
```

```python
sns.stripplot(x = "day", y = "total_bill", data = df, jitter = True);
```

```python
sns.stripplot(x = "tip", y = "day", hue = "sex", data = df, jitter = True);
```

```python
sns.boxplot(x = "day", y = "total_bill", hue = "time", data = df);
```
