```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#encoding='latin1', encoding='iso-8859-1' or encoding='cp1252' found on windows
df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Online_Retail/Online_Retail.csv',encoding='latin1')
df.head()
```

```python
#Create a histogram with the 10 countries that have the most 'Quantity' ordered except UK
df1 = df.groupby('Country').sum().sort_values(by='Quantity',ascending=False)[1:]
df1.head(10)
```

```python
df1['Quantity'].plot(kind='bar')
plt.xlabel('Countries')
plt.ylabel('Quantity')
plt.title('10 Countries with more orders')
plt.show()
```

```python
#Exclude negative Quatity entries
df1[df1['Quantity'] > 0].plot(kind='bar')
plt.xlabel('Countries')
plt.ylabel('Quantity')
plt.title('10 Countries with more orders')
plt.show()
```
