```python
import pandas as pd
import numpy as np
babys = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv")
babys.head(10)
```

```python
del babys['Unnamed: 0']
del babys['Id']
```

```python
babys.head()
```

```python
babys.shape
```

```python
#Is there more male or female names in the dataset?
len(babys['Gender'] == 'M') == len(babys['Gender'] == 'F')
```

```python
#Group the dataset by name and assign to names
names = babys.groupby(['Name']).sum().sort_values(by='Count',ascending=False)
```

```python
names.head()
```

```python
#How many different names exist in the dataset?
len(names.index)
```

```python
#What is the name with most occurrences?
names.head(1)
names.Count.idxmax()
```

```python
#How many different names have the least occurrences?
names[names.Count == names.Count.min()]
```

```python
names[names.Count == names.Count.median()]
```
