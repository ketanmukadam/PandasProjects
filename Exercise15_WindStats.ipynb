```python
import pandas as pd
import numpy as np
import datetime
#parse_dates option combines the listed columns and convert to date
df = pd.read_table("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data",sep='\s+',parse_dates=[[0,1,2]])
df.head()
```

```python
#Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
def fixdate(x):
    if x.year > 2017 :
        return datetime.date(x.year - 100, x.month, x.day)
    else:
        return datetime.date(x.year, x.month, x.day)

df['Yr_Mo_Dy'] = df['Yr_Mo_Dy'].apply(fixdate)
```

```python
# Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns]
df['Yr_Mo_Dy'] = pd.to_datetime(df['Yr_Mo_Dy'])
df.set_index(['Yr_Mo_Dy'],inplace=True)
df.head()
```

```python
#Compute how many values are missing for each location over the entire record
df.isnull().sum()
```

```python
#Compute how many non-missing values there are in total
df.shape[0] - df.isnull().sum()
```

```python
#Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
df.mean()
df.mean().mean()
```

```python
#Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
loc_stat = pd.DataFrame()
loc_stat['min'] = df.min()
loc_stat['max'] = df.max()
loc_stat['mean'] = df.mean()
loc_stat['std'] = df.std()
loc_stat.head()
```

```python
#Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day
day_stat = pd.DataFrame()
day_stat['min'] = df.min(axis=1)
day_stat['max'] = df.max(axis=1)
day_stat['mean'] = df.mean(axis=1)
day_stat['std'] = df.std(axis=1)
day_stat.head()
```

```python
#Find the average windspeed in January for each location
jan_data = df[df.index.month == 1]
jan_data.mean()
```

```python
#Downsample the record to a yearly frequency for each location
df.groupby(df.index.year).mean()
```

```python
# creates a new column 'date' and gets the values from the index
df['date'] = df.index

# creates a column for each value from date
df['month'] = df['date'].apply(lambda date: date.month)
df['year'] = df['date'].apply(lambda date: date.year)
df['day'] = df['date'].apply(lambda date: date.day)
```

```python
#Downsample the record to a yearly frequency for each location
df.query('month == 1 and day == 1')
```

```python
#Downsample the record to a monthly frequency for each location
df.query('day == 1').head()
```

```python
#Calculate the mean windspeed for each month in the dataset
df['monthid'] = df.month * df.year
df.groupby(['monthid']).mean().head()
```

```python
#Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks
df.resample('W').agg(['min','max','mean','std']).iloc[1:53, 0:12*4].head(10)
```
