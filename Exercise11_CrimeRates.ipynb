```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv")
df.head()
```

```python
#What is the type of the columns?
#for x in df.loc[0]:
#    print(type(x))
df.info()
```

```python
#Convert the type of the column Year to datetime64
f = lambda x : np.datetime64(str(x),'Y')
df['Year'] = df['Year'].apply(f)
#df.Year = pd.to_datetime(df.Year, format='%Y')
```

```python
#Set the Year column as the index of the dataframe
df.set_index('Year',inplace=True)
```

```python
df.head()
```

```python
#Delete the Total column
df.drop('Total',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Group the year by decades and sum the values
#Pay attention to the Population column number, summing this column is a mistake
df.set_index((df.index.year//10)*10,inplace=True)

# the resampling method is a nice way to use the date-time index
#cr = df.resample('10AS').sum()
#pop = df['Population'].resample('10AS').max()
#df['Population'] = pop
```
